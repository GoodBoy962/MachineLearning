#### Import libraries for analitics

```python
import numpy as np
import pandas as pa
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Get data about 5 China cities

```python
beijing = pa.read_csv('FiveCitiePMData/beijing.csv')
chengdu = pa.read_csv('FiveCitiePMData/chengdu.csv')
guangzhou = pa.read_csv('FiveCitiePMData/guangzhou.csv')
shanghai = pa.read_csv('FiveCitiePMData/shanghai.csv')
shenyang = pa.read_csv('FiveCitiePMData/shenyang.csv')
```

#### Scaling function

```python
def getScaleValue(data):
    return (data - data.mean())/(data.max() - data.min())
```

#### Function to implement the task

```python
def getDewpAndPressureStats(city):
    
    dewp = city['DEWP']
    pres = city['PRES']
    
    plt.figure(1)
    plt.subplot(2,2,1)
    dewp.plot()
    
    plt.subplot(2,2,2)
    pres.plot()
    
    plt.subplot(2,2,3)
    dewp.plot()
    pres.plot()
    
    dewp_normalized = getScaleValue(dewp)
    pres_normalized = getScaleValue(pres)
    
    plt.subplot(2,2,4)
    dewp_normalized.plot()
    pres_normalized.plot()
    
    print('Correlation between Dew and Pressure is %s' % dewp.corr(pres))
```

### Beijing

```python
getDewpAndPressureStats(beijing)
```

### Chengdu

```python
getDewpAndPressureStats(chengdu)
```

### Guangzhou

```python
getDewpAndPressureStats(guangzhou)
```

### Shanghai

```python
getDewpAndPressureStats(shanghai)
```

### Shenyang

```python
getDewpAndPressureStats(shenyang)
```
