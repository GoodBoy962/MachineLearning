```python
data = pd.read_csv('FiveCitiePMData/chengdu.csv', usecols=[9, 12]).dropna()
```

```python
data.head()
```

```python
plt.plot(data.TEMP, data.DEWP, '.')
```

```python
X = data.TEMP.values.reshape(1, -1)
y = data.DEWP.values.reshape(1, -1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.linear_model import Ridge

alphas = [0.01, 0.1, 1]

for alpha in alphas:
    ridgereg = Ridge(alpha=alpha, normalize=True)
    ridgereg.fit(X, y)
    y_pred = ridgereg.predict(X)
    print(alpha, ridgereg.coef_)
    plt.plot(X, y_pred)
    plt.plot(X, y)
    plt.show()
```
